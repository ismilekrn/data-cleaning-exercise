## Data Cleaning Exercise
In this jupyter notebook I have merged and cleaned data from three separate datasets; Reviews1.csv, Reviews2.csv and Medications.csv. 

```python
import pandas as pd

med=pd.ExcelFile('Medications.xlsx')
med=med.parse("Info")
rev1=pd.read_csv('Reviews1.csv')
rev2=pd.read_csv('Reviews2.csv')
```

```python
#Question 1 - Combining data
revs = [rev1, rev2]
combined_rev = pd.concat(revs)
combined_rev
```

```python
#Question 2 - Cleaning duplicates
combined_rev = combined_rev.drop_duplicates(subset=['Review_ID'])
combined_rev['Drug_name'] = combined_rev['Drug_name'].str.lower() #added this additional code to make 'Drug_name' column lower case. 
combined_rev
```

```python
#Question 3a - Change case
med['Medication'] = med['Medication'].str.lower()
med
```

```python
#Question 3b - Trim whitespaces 
med['Medication'] = med['Medication'].str.strip()
med
```

```python
#Queston 4 - Recode data
med = med.replace(['Rx/OTC'], 'Both')
med
```

```python
#Question 5 - Linking datasets

linked_review = pd.merge(combined_rev, med[['Medication','Rx-otc', 'Csa']], how='left', right_on='Medication', left_on='Drug_name')
linked_review = linked_review.drop(columns=['Medication']) #dropping medication column since it is the same as 'Drug_name' column

#indicator
linked_review
```

```python
#Question 6 - Convert to dummy variables / one hot encoding
#dummy_column = linked_review['Rx-otc']


#linked_review.columns = linked_review.columns.str.strip() #stripping
dummy_variable = pd.get_dummies(linked_review['Rx-otc'])
linked_review['Rx-otc_dummy_variable'] = dummy_variable.cumsum(axis=1).ne(1).sum(axis=1)

linked_review
```

```python
#Question 7 - Working with dates and timestamps
linked_review['Date of Review'] = pd.to_datetime(linked_review['Date of Review'])

start_date = linked_review['Date of Review'] >= '2016-01-01'
end_date = linked_review['Date of Review'] <= '2016-12-31'

linked_review = linked_review.loc[start_date & end_date]
linked_review
```

```python
#Question 8 - Write to csv
linked_review.to_csv('Reviews_Final.csv')
```
